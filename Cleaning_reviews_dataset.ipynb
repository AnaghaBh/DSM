```python
filtered_business = pd.read_csv(r"C:\Users\DELL\Downloads\filtered_business_data.csv")  

file_path = r"C:\Users\DELL\Downloads\final_cleaned_reviews_data.csv"  

df_review = pd.read_csv(file_path)

food_business_ids = set(filtered_business['business_id'])
df_review_food = df_review[df_review['business_id'].isin(food_business_ids)]

print(f"Total reviews for food-related businesses: {len(df_review_food)}")
df_review_food.head()
```

```python
unique_businesses_in_reviews = df_review_food['business_id'].nunique()
print(f"Unique businesses in review dataset: {unique_businesses_in_reviews}")
```

```python
average_reviews_per_business = len(df_review_food) / unique_businesses_in_reviews
print(f"Average reviews per food business: {average_reviews_per_business:.2f}")

duplicates = df_review_food.duplicated(subset=['user_id', 'business_id', 'date'])
print(f"Duplicate review entries: {duplicates.sum()}")

df_review_food = df_review_food.drop_duplicates(subset=['user_id', 'business_id', 'date'])
print(len(df_review_food))
```

```python
business_review_counts = df_review_food['business_id'].value_counts()

valid_business_ids = business_review_counts[(business_review_counts >= 10) & (business_review_counts <= 200)].index

df_review_food = df_review_food[df_review_food['business_id'].isin(valid_business_ids)]

df_review_food['date'] = pd.to_datetime(df_review_food['date'])
df_review_food = df_review_food[df_review_food['date'].dt.year >= 2018]

user_review_counts = df_review_food['user_id'].value_counts()
active_users = user_review_counts[user_review_counts >= 5].index
df_review_food = df_review_food[df_review_food['user_id'].isin(active_users)]
print(len(df_review_food))
```

```python
df = pd.read_csv('cleaned_review_food.csv', quotechar='"', engine='python')

def clean_field(value):
    if isinstance(value, str):
        value = value.replace('"', '""')

        value = value.replace('\n', ' ').replace('\r', '')
       
        if value.startswith('"') and value.endswith('"'):
            value = value[1:-1]
       
        value = ''.join([c if c.isprintable() else ' ' for c in value])

        value = re.sub(r'"[^"]*"', lambda match: match.group(0).strip(), value)

    return value

df = df.applymap(clean_field)

df.to_csv(r"C:\Users\DELL\Downloads\cleaned_review_food_fixed.csv", index=False, quotechar='"', quoting=1)
```

```python
df = pd.read_csv('cleaned_review_food.csv', quotechar='"', engine='python')

def is_problematic_row(value):
    if isinstance(value, str):
        if value.startswith('"') and value.endswith('"'):
            if len(value) > 2 and not value[-1] in ['"', ' ']:
                return True
    return False

problematic_rows = df[df['text'].apply(is_problematic_row)]

print(problematic_rows)
```

```python
file_path = r"C:\Users\DELL\Downloads\cleaned_review_food.csv"
checkin_df = pd.read_csv(file_path)

checkin_df['text'] = checkin_df['text'].fillna('')

def clean_text(text):
    text = str(text)
    text = text.replace('\n', ' ')       # remove newlines
    text = text.replace('\r', ' ')
    text = text.replace('"', '')         # remove double quotes
    text = text.replace("'", '')         # remove single quotes
    text = text.replace("\\", '')        # remove backslashes
    text = text.replace("=", '')         # remove equal signs
    text = re.sub(r'[^\x00-\x7F]+', '', text)  # remove non-ASCII characters
    return text.strip()

checkin_df['text'] = checkin_df['text'].apply(clean_text)

output_path = r"C:\Users\DELL\Downloads\cleaned_checkin_for_neo4j.csv"
checkin_df.to_csv(output_path, index=False)
print("Cleaned file saved successfully.")
```
