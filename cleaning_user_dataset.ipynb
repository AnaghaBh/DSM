```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = r"C:\Users\DELL\Downloads\final_cleaned_user_data.csv" 
df = pd.read_csv(file_path)

print("Head of the dataset:")
display(df.head())

print("Info:")
df.info()

print("\nSummary statistics:")
display(df.describe())

if 'review_count' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(df['review_count'], bins=50, kde=True)
    plt.title('Distribution of Review Count')
    plt.xlabel('Review Count')
    plt.ylabel('Frequency')
    plt.show()
```

```python
df = df.dropna(subset=['user_id'])

df['elite'] = df['elite'].fillna('')
df['friends'] = df['friends'].fillna('')

df = df[df['review_count'] > 0]

print(f"Remaining users after filtering: {len(df)}")
df.head()
```

```python
df = df[~((df['review_count'] < 5) & (df['friend_count'] == 0))]
df.head()
print(len(df))
df.to_csv(r"C:\Users\DELL\Downloads\prematched_user_data.csv", index=False)

```
