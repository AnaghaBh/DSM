```python
df = pd.read_csv(r"C:\Users\DELL\Downloads\final_cleaned_business_data.csv")  

unique_names = df['categories'].nunique()
total_names = len(df['categories'])

print(f"Unique catgory names: {unique_names}")
print(f"Total rows: {total_names}")
```

```python
import ast
df = df.dropna(subset=['categories'])

def parse_categories(val):
    try:
        parsed = ast.literal_eval(val)
        if isinstance(parsed, list):
            return parsed
        else:
            return [val]  
    except:
        return [val]  

df['categories'] = df['categories'].apply(parse_categories)

df['category_tuple'] = df['categories'].apply(lambda x: tuple(sorted(x)))

unique_category_lists = df['category_tuple'].drop_duplicates()
```
